```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
ne=60
nangle=3
nphi=128

t_res=1/64

t_seg=int(nphi*t_res)

fp1=np.zeros((ne,nangle,nphi))
dp1=np.genfromtxt('comp1.dat')
fp2=np.zeros((ne,nangle,nphi))
dp2=np.genfromtxt('comp2.dat')

for i in range(ne):
    id=0
    for j in range(nangle):
        for k in range(nphi):
            fp1[i][j][k]=dp1[i,3*id+1]#与输出结构有关
            fp2[i][j][k]=dp2[i,3*id+1]

            id=id+1 
```

```python
viewer=2
counts1=np.zeros(nphi)
counts2=np.zeros(nphi)

for i in range(nphi):
    counts1[i]=np.sum(fp1[:,viewer,i])
    counts2[i]=np.sum(fp2[:,viewer,i])
time=np.linspace(0,t_seg,nphi)
```

```python
fig,ax1=plt.subplots()
ax1.plot(time,counts1)

fig,ax2=plt.subplots()
ax2.plot(time,counts2)
```
